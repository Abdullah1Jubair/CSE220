```python
#For Task 1
class Node:
    def __init__(self, val):
        self.elem = val
        self.next = None

adjList = np.full(10, None)

def list_graph(adjList, n1, n2):
    temp = adjList[n1]
    newNode = Node(n2)

    if temp is None:
        adjList[n1] = newNode
    else:
        while temp.next is not None:
            temp = temp.next
        temp.next = newNode
    temp = adjList[n2]
    newNode = Node(n1)

    if temp is None:
        adjList[n2] = newNode
    else:
        while temp.next is not None:
            temp = temp.next
        temp.next = newNode
def degree(adjList):
    max_degree = 0
    for i in range(len(adjList)):
        count = 0
        temp = adjList[i]
        while temp is not None:
            count += 1
            temp = temp.next
        if count > max_degree:
            max_degree = count
    return max_degree

def showAdjList(adjList):
    def showList(adjList, n1):
        temp = adjList[n1]
        print(n1, end=': ')
        while temp is not None:
            if temp.next is not None:
                print(f'{temp.elem}', end=' -> ')
            else:
                print(f'{temp.elem}', end='')
            temp = temp.next
        print()

    for indx in range(len(adjList)):
        showList(adjList, indx)

list_graph(adjList, 1, 2)
list_graph(adjList, 1, 3)
list_graph(adjList, 2, 3)
list_graph(adjList, 2, 4)
list_graph(adjList, 2, 5)
list_graph(adjList, 3, 4)
list_graph(adjList, 3, 5)
list_graph(adjList, 3, 6)
list_graph(adjList, 4, 6)
list_graph(adjList, 4, 5)
list_graph(adjList, 5, 6)
list_graph(adjList, 7, 8)
list_graph(adjList, 7, 9)
list_graph(adjList, 8, 9)
list_graph(adjList, 8, 6)
list_graph(adjList, 6, 7)
showAdjList(adjList)
print("-----------------------")
print("Maximum degree is ",degree(adjList))
```

```python
import numpy as np

class Node:
    def __init__(self, val, weight=None):
        self.elem = val
        self.weight = weight
        self.next = None
def list_graph(adjList, n1, n2, weight):
    temp = adjList[n1]
    newNode = Node(n2, weight)
    if temp is None:
        adjList[n1] = newNode
    else:
        while temp.next is not None:
            temp = temp.next
        temp.next = newNode

def convert_to_undirected(adjList):
    undirected_adjList = np.full(len(adjList), None)

    for n1 in range(1, len(adjList)):
        temp = adjList[n1]
        while temp is not None:
            n2 = temp.elem
            weight = temp.weight

            list_graph(undirected_adjList, n1, n2, weight)
            list_graph(undirected_adjList, n2, n1, weight)

            temp = temp.next

    return undirected_adjList
def showAdjList(adjList):
    def showList(adjList, n1):
        temp = adjList[n1]
        print(n1, end=': ')
        while temp is not None:
            if temp.next is not None:
                print(f'({temp.elem},{temp.weight})', end=' -> ')
            else:
                print(f'({temp.elem},{temp.weight})', end='')
            temp = temp.next
        print()

    for indx in range(1, len(adjList)):
        showList(adjList, indx)

adjList = np.full(10, None)

list_graph(adjList, 1, 2, 5)
list_graph(adjList, 1, 3, 10)
list_graph(adjList, 2, 3, 3)
list_graph(adjList, 2, 4, 7)
list_graph(adjList, 2, 5, 2)
list_graph(adjList, 3, 4, 8)
list_graph(adjList, 3, 5, 4)
list_graph(adjList, 3, 6, 6)
list_graph(adjList, 4, 6, 1)
list_graph(adjList, 4, 5, 9)
list_graph(adjList, 5, 6, 2)
list_graph(adjList, 7, 8, 3)
list_graph(adjList, 7, 9, 11)
list_graph(adjList, 8, 9, 4)
list_graph(adjList, 8, 6, 12)
list_graph(adjList, 6, 7, 5)

print("Directed Weighted Graph:")
showAdjList(adjList)

undirected_adjList = convert_to_undirected(adjList)

print("\nUndirected Weighted Graph:")
showAdjList(undirected_adjList)
```
